```python
# Does line execute?                        Yes    No
#                                           ---    --
if 'foo' in ['foo', 'bar', 'baz']:        #  x
    print('Outer condition is true')      #  x

    if 10 > 20:                           #  x
        print('Inner condition 1')        #        x

    print('Between inner conditions')     #  x

    if 10 < 20:                           #  x
        print('Inner condition 2')        #  x

    print('End of outer condition')       #  x
print('After outer condition')            #  x
```

```python
# Does line execute?                        Yes    No
#                                           ---    --
if 'foo' in ['foo', 'bar', 'baz']:        #  x
    print('Outer condition is true')      #  x

    if 10 > 20:                           #  x
        print('Inner condition 1')        #        x
        
    print('Between inner conditions')     #  x

    if 10 < 20:                           #  x
        print('Inner condition 2')        #  x
        print('30')
        
    print('End of outer condition')       #  x
print('After outer condition')            #  x
```

```python
# while loops
n = 5
while n > 0:    # 5 4 3 2 1 0>0?? False jadi stop loopnya 
    n -= 1      # 4 3 2 1 0         #bacanya ke bawah ya bestiee
    print(n)    # 4 3 2 1 0
```

```python
i = 1
while i < 6 :  # 1 2 3 4 5
    print(i)   # 1 2 3 4 5 # ini yang diprint yaa
    i += 1     # 2 3 4 5 6 6<6?? false
```

```python
n = 5                              
while n > 0:                       # 5      4      3
    n -= 1                         # 4      3      2
    if n == 2:                     # false  false  false
        break # Break Statement    # skip   skip   breakkk
    print(n)                       # 4      3      
print('Loop ended.')
```

```python
n = 5                 
while n > 0:          # 5 4 3 2 1 0>0?? ended
    n -= 1            # 4 3 2 1 0
    if n == 2:        # f f t f f
        continue      # - - continue
    print(n)          # 4 3 - 1 0
print('Loop ended.')
```

```python
#nested while loops
age = 60 
gender = 'M'

if age < 18:
    if gender == 'M':
        print('son')
    else:
        print('daughter')
elif age >= 18 and age < 65:
    if gender == 'M':
        print('father')
    else:
        print('mother')
else:
    if gender == 'M':
        print('grandfather')
    else:
        print('grandmother')
```
