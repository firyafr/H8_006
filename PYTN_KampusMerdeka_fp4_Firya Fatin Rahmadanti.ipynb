## Clustering 

Clustering merupakan salah satu algoritma yang bekerja dengan cara mengelompokkan obyek-obyek data (pola, entitas, kejadian, unit,hasil observasi) ke dalam cluster tertentu (Xu and Wunsch,2009). Dengan kata lain algoritma Clustering melakukan pemisahan/ pemecahan/ segmentasi data ke dalam sejumlah kelompok (cluster) menurut karakteristik tertentu. Dalam klasifikasi unsupervised (clustering/segmentation/partitioning) menggunakan data yang tidak memilki label/Target seperti pada klasifikasi supervised, namun dari atribut yang ada akan dikelompokkan menurut karakteristiknya. 

Dalam prakteknya, clustering membantu mengidentifikasi dua kualitas data:
- Meaningfull, Meaningful clusters expand domain knowledge. hanya sebagai proses awal untuk kemudian dilanjutkan dengan pekerjaan seperti summarization(rata-rata,standar deviasi), pelabelan kelas untuk setiap kelompok sehingga dapat digunakan sebagai data training dalam klasifikasi supervised.
- usefull, Useful clusters berfungsi sebagai intermediate step pada data pipeline*.

Clustering dapat dibedakan menurut struktur kelompok ,keanggotaan data dalam kelompok, dan kekompakan data dalam kelompok. 
Menurut struktur kelompok clustering dibagi menjadi dua yaitu hierarchical dan partitioning. 
- HIERARCHICAL:menemukan klaster yang berurutan menggunakan klaster yang telah ditetapkan sebelumnya
    • aglomerative (bottom-up): mulai dengan setiap elemen dalam cluster terpisah dan gabungkan mereka sesuai dengan properti yang diberikan. 
    • divisive (top‐down) : merupakan proses kebalikan dari agglomerative
    
- PARTITIONAL: biasanya menentukan semua cluster sekaligus.mengelompokkan datapointkedalam k klaster tanpa struktur hirarki( Xu & Wunsch, 2009). Partitioning algorithm memisahkan (mem-partisi) titik-titik atau objek data menjadi K kelompok/ cluster, sedemikian rupa sehingga titik-titik dalam satu cluster memiliki karakteristik se-homogen mungkin, dan karakteristik antar cluster se-heterogen mungkin. Beberapa algoritma yang termasuk dalam metode ini adalah K-Means, K-Medoid, DBSCAN, Mixture-Based Density,Graph Theory-Based Clustering,Fuzzy Clustering.

Dalam menentukan jumlah cluster dilakukan menggunakan Metode Elbow dan Shilouette Score.

#### Elbow Method
Metode Elbow merupakan suatu metode yang digunakan untuk menghasilkan informasi dalam menentukan jumlah cluster terbaik dengan cara melihat persentase hasil perbandingan antara jumlah cluster yang akan membentuk siku pada suatu titik.

#### Shilouette Score
Metode Shilouette merupakan metode evaluasi cluster yang menggabungkan metode cohessian dan separation. Cohessian diukur dengan menghitung seluruh objek yang terdapat dalam sebuah cluster dan separation diukur dengan menghitung jarak rata-rata setiap objek dalam sebuah cluster dengan cluster terdekatnya. 


**Dalam Analisis kali ini akan dilakukan clustering menggunakan algoritma K-Means, PCA, K-Medoid dan DBSCAN**



### K-Means Clustering

Dalam machine-learning dan statistic K-Means merupakan metode analisis kelompok yang mengarah pada pembagian N obyek pengamatan ke dalam K kelompok (cluster), dimana setiap obyek dimiliki oleh sebuah kelompok dengan mean (rata-rata) dan metode ini mencoba untuk menemukan pusat dari kelompok (centroid) dalam data sebanyak iterasi perbaikan yang dilakukan. Metode ini berusaha
membagi data kedalam kelompok sehingga data yang berkarakteristik sama dimasukkan ke dalam satu kelompok sementara data yang berkarakteristik berbeda dimasukkan dalam kelompok yang lain. 


### PCA (Principal Component Analysis)
   Principal Component Analysis (PCA) adalah teknik *unsupervised Machine Learning* yang mencoba menurunkan serangkaian fitur berdimensi rendah dari rangkaian yang jauh lebih besar sambil tetap mempertahankan varians sebanyak mungkin. PCA sangat berguna untuk menyimpan informasi sekaligus mereduksi dimensi data. Namun, kita perlu memastikan bahwa data sudah diskalakan (scaled) dengan benar untuk mendapatkan PCA yang berguna.

### K-Medoid Clustering
K-Medoids atau Partitioning Around Method (PAM) adalah metode cluster non hirarki yang merupakan varian dari metode K-Means.
K-Medoids hadir untuk mengatasi kelemahan K-Means yang sensitif terhadap outlier karena suatu objek dengan suatu nilai yang besar mungkin secara substansial menyimpang dari distribusi data (Jiawei & Kamber, 2006). Hal ini didasarkan pada penggunaan medoids bukan dari pengamatan mean yang dimiliki oleh setiap cluster, dengan tujuan mengurangi sensitivitas dari partisi sehubungan dengan nilai ekstrim yang ada dalam dataset (Vercellis, 2009)

### DBSCAN
Algoritma DBSCAN adalah sebuah algoritma clustering yang dikembangkan berdasarkan tingkat kerapatan data (density-based). Dimana algoritma ini menumbuhkan daerah yang memiliki kerapatan tinggi menjadi cluster-cluster, dan menemukan cluster-cluster tersebut pada bentuk bebas dalam sebuah ruang database dengan memanfaatkan noise. DBSCAN sering diterapkan pada data yang banyak mengandung noise, hal ini dikarenakan DBSCAN tidak akan memasukkan data yang dianggap noise kedalam cluster manapun.

DBSCAN memerlukan dua parameter input sebelum melakukan proses clustering yaitu epsilon (eps) dan minimum points (minPts). Epsilon merupakan jarak maksimal antara dua data dalam satu cluster yang diizinkan, dan minimum points adalah banyaknya data minimal dalam jarak epsilon agar terbentuk suatu cluster. Metode jarak yang digunakan dalam DBSCAN adalah jarak Euclidian.

### Transformasi data
Periksa the skewed values and transform them

```python
float_columns = [x for x in df.columns if x not in  ['CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'TENURE']]
```

```python
skew_columns = (df[float_columns].skew().sort_values(ascending=False))
skew_columns = skew_columns.loc[skew_columns > 0.75]
skew_columns
```

```python
# Perform a Log transform on skewed columns
for col in skew_columns.index.tolist():
    df[col] = np.log1p(df[col])
```

Akan dilakukan standarisasi menggunakan StandardScaler.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Standardizing the data
scaler = StandardScaler()
scaler.fit(df)
scaled_features = scaler.transform(df)
data_scaled = pd.DataFrame(scaled_features,columns = df.columns)
data_scaled.head()
```

## Menentukan Jumlah Cluster Menggunakan Metode Elbow dan Shilouette Score

```python
from sklearn.cluster import KMeans
```

### Kurva Elbow

```python
# Membuat plot kurva Elbow
scores_1 = []

range_values = range(1,10)
for i in range_values:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_scaled)
    scores_1.append(kmeans.inertia_)
plt.plot(scores_1, 'bx-')
plt.style.use('ggplot')
plt.title('Kurva Elbow')
plt.xlabel('Numer of Clusters')
plt.ylabel('Scores') 
plt.show()
```

**Interpretasi:** Berdasarkan kurva di atas, jumlah cluster yang palig optimal adalah 2 karena titk 2 membentuk siku kurva

### Shiloutte Score

```python
from sklearn.metrics import silhouette_score
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice we start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans_3 = KMeans(n_clusters=k)
    kmeans_3.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans_3.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.title('Kurva Shilloutte Score')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

**Interpretasi:** Jumlah cluster yang paling optimal berdasarkan shilloutte score adalah 2, karena 2 merupakan nilai yang paling tinggi
